```python
import pandas as pd
import numpy as np
import xgboost as xgb
import os
import pickle
```

#30 Seconds Dataset

```python
path_to_data = '../datasets/30_sec_datasets/xgboost_datasets'
X_train = pickle.load(open(os.path.join(path_to_data, 'X_train.pkl'), 'rb'))
y_train = pickle.load(open(os.path.join(path_to_data, 'y_train.pkl'), 'rb'))
X_test = pickle.load(open(os.path.join(path_to_data, 'X_test.pkl'), 'rb'))
y_test = pickle.load(open(os.path.join(path_to_data, 'y_test.pkl'), 'rb'))
X_eval = pickle.load(open(os.path.join(path_to_data, 'X_eval.pkl'), 'rb'))
y_eval = pickle.load(open(os.path.join(path_to_data, 'y_eval.pkl'), 'rb'))
```

```python
def train_xgboost_model(X_train, y_train, X_eval, y_eval, batch_size):
    dtrain = xgb.DMatrix(X_train, label=y_train)
    deval = xgb.DMatrix(X_eval, label=y_eval)
    params = {
        'objective': 'multi:softprob',  # Change to 'binary:logistic' if you have a binary classification problem
        'num_class': len(set(y_train)),  # Number of unique classes
        'eval_metric': 'mlogloss',  # Change to 'logloss' for binary classification
        'learning_rate': 0.1,
        'max_depth': 6,
        'min_child_weight': 1,
        'subsample': 0.8,
        'colsample_bytree': 0.8,
        'n_estimators': 1000,  # Number of trees to build
        'seed': 42
    }

    evals = [(dtrain, 'train'), (deval, 'eval')]
    evals_result = {}  # Dictionary to store evaluation results
    model = xgb.train(params, dtrain, num_boost_round=1000, evals=evals, early_stopping_rounds=10, evals_result=evals_result)

    history = {'history': {'val_loss': evals_result['eval']['mlogloss']}}

    return model, history


model, history = train_xgboost_model(X_train, y_train, X_eval, y_eval, batch_size=64)
```

```python
y_test_pred = model.predict(xgb.DMatrix(X_test))
pickle.dump(y_test_pred, open('test_classifications/30_sec_xgboost_predictions.pkl', 'wb'))
```
